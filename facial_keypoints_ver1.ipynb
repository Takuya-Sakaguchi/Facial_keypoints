#  here I am coverting the x and y corrdiantes of the left eye center postion to one value

```python
def convertXYtoPosition(x, y):
    return int(y)*96 + int(x)
```

```python
def convertRecordsToPosition(i):
    if not np.isnan(df_train.left_eye_center_x[i]):
        if not np.isnan(df_train.left_eye_center_y[i]):
            return convertXYtoPosition(df_train.left_eye_center_x[i],df_train.left_eye_center_y[i])
    else:
        return 0
```

```python
convertRecordsToPosition(2)
```

```python
position = []
for i in range(len(df_train)):
    position.append(convertRecordsToPosition(i))
    
    
```

```python
position
```

```python
postion = np.array(position)
```

```python
y_train = position
```

# Here I simply try the random forest first

```python
from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
from sklearn.ensemble import RandomForestClassifier
Xtrain, Xtest, ytrain, ytest = train_test_split(x_train, y_train, random_state=0)
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(Xtrain, ytrain)
ypred = clf.predict(Xtest)
print(metrics.accuracy_score(ypred, ytest))
```
